```python
import seaborn as sns
```

```python
df = pd.read_csv('../Dataset/x_train.csv')
```

```python
output_df = pd.read_csv('../Dataset/y_train.csv')
```

```python
output_df.head()
```

## Concatenating Features and Output Variable Data frames

```python
df_copy = pd.concat([df, output_df], axis=1).reindex(df.index)
```

```python
df_copy.head()
```

```python
missing_variables = df_copy.isnull().sum()
missing_values = missing_variables[missing_variables > 0]
```

```python
print(missing_values)
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df_copy['_MICHD'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Disease')
ax[0].set_ylabel('')
sns.countplot(x='_MICHD',data=df_copy,ax=ax[1])
ax[1].set_title('Disease')
plt.show()
```

```python
value_counts = df_copy['CADULT'].value_counts()

# Extract the categories (unique values) and their counts
categories = value_counts.index
counts = value_counts.values
print(f'Counts: {counts}')

# Create a bar plot
plt.bar(categories, counts)

# Set custom labels for the x-axis ticks
plt.xticks(categories, ['Male (1)', 'Female (2)'])

# Label the axes and set the title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of CADULT Categories')
```

### Comparing Gender to Disease Positivity

```python
df_copy.groupby(['CADULT','_MICHD'])['_MICHD'].count()
```

```python
df_copy[['CADULT', '_MICHD']]
```

```python
fig, ax = plt.subplots(1,2, figsize = (18, 8))
df_copy[['CADULT', '_MICHD']].groupby(['CADULT']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Gender Vs Disease')

sns.countplot(x='CADULT', hue='_MICHD', data= df_copy, ax=ax[1])
ax[1].set_title('Gender Vs Count')
```

It can be seen that the Males have more heart disease postive cases compared to the Female Gender
